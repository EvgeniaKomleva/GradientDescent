```python
%matplotlib inline
import matplotlib.pyplot as plt
from time import time
import sympy as sp
import numpy as np
import math as mt
import numpy.linalg as la
from scipy.misc import derivative

size_0 = 100 #размерность пространства
size  = size_0
e = 1e-1

iterations = 15

time1 = np.zeros(iterations)
time2 = np.zeros(iterations)

k1 = np.zeros(iterations)
k2 = np.zeros(iterations)

e1 = np.zeros(iterations)
e2 = np.zeros(iterations)

k = 0
#функция вычисления расстояния
def distance(y, z):
    global size
    subdistance  = 0
    for i in range(0, size):
        subdistance = subdistance + (y[i] - z[i])*(y[i] - z[i])
    distance = mt.sqrt(subdistance)
    return distance

#слагаемые исследуемых функций
def f_1(x):
    return (mt.exp(x) - x)/10

def f_2(x):
    return mt.log(mt.exp(x) + mt.exp(-x))

#исследуемые функции
def function_1(y):
    global size
    funct = 0
    for i in range(0, size):
        funct = funct + (i + 1) * f_1(y[i])
    return funct
        
def function_2(y):
    global size
    funct = 0
    for i in range(0, size):
        funct = funct + f_2(y[i])
    return funct


while size < size_0 * (iterations + 1):
    gradient = np.ones(size) 
    gessian = np.zeros((size,size)) 
    invgessian = np.zeros((size,size)) 

    y = np.ones(size)
    z = np.zeros(size)
    p = np.zeros(size)

    epsilon = 0.01
    alpha = 100
    theta = 0.7
    k1[k] = 0
    #градиентный спуск для функции 1
    begin1 = time()
    while distance(y,z) > e: #критерий остановки - 1
    #while abs(function_1(y) - function_1(z)) > e: #критерий остановки - 2
    #while distance(gradient, np.zeros(size)) > e:

        #поиск градиента
        for i in range(0, size):
            gradient[i] = (i + 1) * derivative(f_1, y[i], dx=1e-6)

        #поиск матрицы вторых производных
        for i in range(0, size):
            gessian[i, i] = (i + 1) * derivative(f_1, y[i], dx=1e-6, n = 2)

        invgessian = la.inv(gessian) #обратная матрица к матрице Гесса
        p = - invgessian.dot(gradient) #направление градиентного спуска

        #одномерная оптимизация - метод Армихо
        while function_1(y + alpha * p) - function_1(y) > epsilon * np.dot(gradient, p):
            alpha = theta * alpha
        z = y
        y = y + alpha * p #шаг градиентного спуска
        k1[k] = k1[k] + 1
    end1 = time()
    time1[k] = end1 - begin1


    gradient = np.ones(size)
    gessian = np.zeros((size,size))
    invgessian = np.zeros((size,size))

    y = np.ones(size)
    z = np.zeros(size)
    p = np.zeros(size)

    alpha = 100
    k2[k] = 0
    begin2 = time()
    #градиентный спуск для функции 2
    while distance(y,z) > e:#критерий остановки - 1
    #while abs(function_1(y) - function_1(z)) > e: #критерий остановки - 2
    #while distance(gradient, np.zeros(size)) > e:
        #поиск градиента
        for i in range(0, size):
            gradient[i] = derivative(f_2, y[i], dx=1e-6)


        #поиск матрицы вторых производных
        for i in range(0, size):
            gessian[i, i] = derivative(f_2, y[i], dx=1e-6, n = 2)

        invgessian = la.inv(gessian) #обратная матрица к матрице Гесса
        p = - invgessian.dot(gradient) #направление градиентного спуска

        #одномерная оптимизация - метод Армихо
        while function_1(y + alpha * p) - function_1(y) > epsilon * np.dot(gradient, p):
            alpha = theta * alpha
        z = y
        y = y + alpha * p #шаг градиентного спуска
        k2[k] = k2[k] + 1
  
    end2 = time()  
    time2[k] = end2 - begin2
    k = k + 1
    size = size + size_0

plt.figure(figsize = (12,8))
axixx = np.arange(size_0, size_0 * (iterations + 1), size_0)
plt.plot(axixx, time1, c = 'blue', label = 'function_1')
plt.plot(axixx, time2, c = 'red', label = 'function_2')
plt.legend() 
plt.grid() 
plt.xlabel('size') 
plt.ylabel('time')
plt.title('График зависимости времени от размерности пространства')
```
